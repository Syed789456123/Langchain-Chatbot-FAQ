Exercise 1: Use the UnstructuredPDFLoader to load the FAQ PDF from the documents folder. Use the load method. (Use default settings)


```python
from langchain_community.document_loaders.pdf import UnstructuredPDFLoader
#from langchain.document_loaders import PyPDFLoader
# Instantiate the UnstructuredPDFLoader
file_path = 'us_en_faqs.pdf'
loader = UnstructuredPDFLoader(file_path=file_path)

# Load the PDF file
docs = loader.load()
```

```python
docs
```

```python
from langchain.schema import Document
from langchain.text_splitter import CharacterTextSplitter

# Convert the list of chunks into a list of Document objects
#documents = [Document(page_content=chunk) for chunk in chunks]

# Split the documents into chunks using the CharacterTextSplitter
split_text = CharacterTextSplitter(separator="\n\n", chunk_size=100, chunk_overlap=10)
split_chunks = split_text.split_documents(docs)
```

```python
split_chunks
```

## We use embeddings provided by HuggingFace.**bold text**

```python
!pip install sentence-transformers
from langchain.embeddings import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings()
```

```python
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceInstructEmbeddings
!pip install InstructorEmbedding
!pip install faiss-cpu
!pip install faiss-gpu

instructor_embeddings = HuggingFaceInstructEmbeddings(model_name="TinyLlama/TinyLlama-1.1B-Chat-v1.0")

```

```python
documents = split_chunks
db_instructEmbedd = FAISS.from_documents(documents, instructor_embeddings)

retriever = db_instructEmbedd.as_retriever(search_kwargs={"k": 3})
```

```python
# Getting LLM from HuggingFaceHub
llm = HuggingFaceHub(repo_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0", model_kwargs={"temperature":0.7, "max_length":256})
```

```python
from langchain.chains.question_answering import load_qa_chain
```

```python
# Creating a load_qa_chain
chain = load_qa_chain(llm, chain_type="stuff", verbose=True)
# Chain type stuff means all the top n chunks
# searched by vector store are stuffed into the prompt
```

```python


# Assuming you have already initialized document_chunks and embeddings
# Create a VectorStore instance using FAISS
faiss_store = FAISS.from_documents(split_chunks, embeddings)
query = "What is Cash-on-Delivery?"
# Search for similar documents to the query using the similarity_search method
docs = faiss_store.similarity_search(query, k=3)
print(docs)
```

```python
#from langchain.vectorstores import FAISS

# Assuming you have already initialized document_chunks and embeddings
# Create a VectorStore instance using FAISS
#faiss_store = FAISS.from_documents(split_chunks, embeddings)

# Search for similar documents to the query using the similarity_search method
docs = faiss_store.similarity_search(query, k=3)
print(docs)
```

```python
# Stuffing the documents and
# inputting the query into the default QnA template
chain.run(input_documents=docs, question=query)
```

```python
from langchain.chains import RetrievalQA
from langchain.vectorstores.base import VectorStoreRetriever
vectorstore_retriever = VectorStoreRetriever(vectorstore=faiss_store, search_kwargs={"k": 3})
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vectorstore_retriever)
```

```python
query = "Tell me about Aramex Freight."
qa_chain.run(query)
```

```python
from langchain.memory import ConversationBufferWindowMemory
from langchain.chains import ConversationalRetrievalChain


memory = ConversationBufferWindowMemory(k=2, memory_key="chat_history", return_messages=True)
```

```python
conversation_chain=ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=retriever,
    memory=memory,
    verbose=True
)
```
